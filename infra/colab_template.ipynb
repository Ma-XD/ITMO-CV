Запусти эту ячейку **в начале каждой Colab-сессии**, чтобы подтянуть свежий код. Потом открой нужный лаб-ноутбук (например `lab1-CLAS/lab1.ipynb`) — он самодостаточный, дальше всё сделает сам.

Перед первым запуском: Tools → Secrets → `GITHUB_TOKEN` (PAT с правом `repo:read` на репо).


```python
import os
from google.colab import userdata

TOKEN = userdata.get("GITHUB_TOKEN")
REPO = "Ma-XD/ITMO-CV"
REPO_DIR = "/content/ITMO-CV"

if os.path.exists(REPO_DIR):
    !cd {REPO_DIR} && git pull
else:
    !git clone https://{TOKEN}@github.com/{REPO}.git {REPO_DIR}
print(f"📂 {REPO_DIR}")
```
