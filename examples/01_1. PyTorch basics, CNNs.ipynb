#### Создадим свою дифференцируемую функцию

```python
class MyStrangeMul(torch.autograd.Function):
    @staticmethod
    def forward(ctx,
                tensor1: torch.FloatTensor,
                tensor2: torch.FloatTensor) -> torch.FloatTensor:
        print(torch.is_grad_enabled())
        ctx.tensor1 = tensor1
        ctx.tensor2 = tensor2
        return tensor1 * tensor2

    @staticmethod
    def backward(ctx, grad: torch.FloatTensor) -> torch.FloatTensor:
        print(torch.is_grad_enabled())
        return grad * ctx.tensor2 * 2, grad * ctx.tensor1
```

```python
x = torch.tensor(10.0, requires_grad=True)
y = torch.tensor(2.0, requires_grad=True)
z = MyStrangeMul.apply(x, y)
z.backward()
x.grad, y.grad
```

```python
x = torch.tensor(10.0, requires_grad=True)
z = MyStrangeMul.apply(x, x)
z.backward()
x.grad
```

#### BatchNorm

##### Custom BatchNorm1d

```python
features_cnt = 15
input = torch.randn(4, features_cnt) * 100
input
```

```python
class CustomBatchNorm1d(nn.Module):
    def __init__(self, num_features, eps=1e-5, momentum=0.1):
        super(CustomBatchNorm1d, self).__init__()
        self.x_norm = None
        self.sqrtvar = None
        self.xmu = None
        self.EX = 0
        self.VarX = 1
        self.eps = eps
        self.momentum = momentum
        self.bias = nn.Parameter(torch.zeros(num_features))  # beta
        self.weight = nn.Parameter(torch.ones(num_features))  # gamma

    def forward(self, x):
        if self.training:
            mean = x.mean(axis=0)
            var = x.var(axis=0, correction=0)
            # подсчет статистики:
            var_corrected = x.var(axis=0, correction=1)
            self.EX = self.EX * (1 - self.momentum) + mean * self.momentum
            self.VarX = self.VarX * (1 - self.momentum) + var_corrected * self.momentum
        else:
            mean = self.EX
            var = self.VarX

        self.xmu = x - mean
        self.sqrtvar = torch.sqrt(var + self.eps)
        self.x_norm = self.xmu / self.sqrtvar
        return self.x_norm * self.weight + self.bias
```

```python
cbn = CustomBatchNorm1d(features_cnt)

output = cbn(input)
output
```

```python
cbn.EX, cbn.VarX
```

#### Создадим сеть с линейными слоями

```python
input_size = 28 * 28
output_size = 10
```

```python
class Net(nn.Module):
    def __init__(self, input_size, output_size):
        super(Net, self).__init__()
        self.l1 = nn.Linear(input_size, 512)
        self.l2 = nn.Linear(512, 128)
        self.l3 = nn.Linear(128, output_size)
        self.act = nn.Tanh()

    def forward(self, x):
        x = self.l1(x)
        x = self.act(x)
        x = self.l2(x)
        x = self.act(x)
        x = self.l3(x)
        return x

net = Net(input_size, output_size)
net
```

```python
def init_weights(m):
    if isinstance(m, nn.Linear):
        nn.init.xavier_uniform_(m.weight)
        m.bias.data.fill_(0.01)

net.apply(init_weights)
```

```python
for p in net.parameters():
  print(p.shape, p.requires_grad)
```

```python
for p in net.buffers():
  print(p.shape, p.requires_grad)
```

#### Критерий и оптимизатор

```python
criterion = nn.CrossEntropyLoss()
```

```python
from torch import optim

optimizer = optim.SGD(net.parameters(), lr=0.05)
```

#### Обучаем и тестируем

```python
device = 'cuda'

def train_one_epoch():
    net.train()
    losses = []
    for i, batch in enumerate(train_dataloader):
        x, labels = batch

        x = x.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        x = x.flatten(start_dim=1)
        out = net(x)

        loss = criterion(out, labels)
        loss.backward()

        if i % 50 == 0:
            losses.append(loss.detach().cpu())

        optimizer.step()

        if i % 200 == 0:
            print(f"Step: {i}/{len(train_dataloader)}. Loss: {loss.detach():.3f}")

    plt.plot(losses)
    plt.show()
```

```python
net.to(device)
print(net.l1.weight.device)

train_one_epoch()
```

```python
import torch
from tqdm import tqdm
import numpy as np

def run_test():
  net.eval()
  losses = []
  with torch.no_grad():
      for data in tqdm(test_dataloader):
          x, labels = data
          x = x.to(device)
          labels = labels.to(device)

          x = x.flatten(start_dim=1)

          out = net(x)

          loss = criterion(out, labels)
          losses.append(loss.detach().cpu())

  return np.mean(losses)
```

```python
run_test()
```

# Свертка

Семейство ConvXd:
* [**Conv1d**](https://pytorch.org/docs/stable/generated/torch.nn.Conv1d.html) - одномерная свертка (одномерное ядро). Предназначется для обработки последовательностей (текст, временные ряды и т.п.), где вход имеет одно измерение.
* [**Conv2d**](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html) - двумерная свертка (двумерное ядро). Предназначется для обработки изображений, где вход имеет два измерения (ширина и высота).
* [**Conv3d**](https://pytorch.org/docs/stable/generated/torch.nn.Conv3d.html) - трехмерная свертка (трехмерное ядро). Предназначется для обработки трехмерных данных, таких как объемные данные или видеокадры, где вход имеет три пространственных измерения (ширина, высота и глубина).

*в этом тексте ядра подразумеваются без учета размерности канала

<img src="https://github.com/IzhanVarsky/useful_images/blob/main/images/appendix_C_eg_3.png?raw=true" alt="drawing" width="700"/>

```python
def get_out_size(input_size, kernel_size, stride_size=1, padding_size=0):
  return (input_size + 2 * padding_size - kernel_size) // stride_size + 1
```

### Напишем свою свертку

```python
import math

_pair = lambda x: x if isinstance(x, tuple) else (x, x)

class BrokenConv2d(nn.Module):
    def __init__(self,
                 in_channels: int,
                 out_channels: int,
                 kernel_size,
                 stride=1,
                 padding=0):
        super(BrokenConv2d, self).__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels

        self.kernel_size = _pair(kernel_size)
        self.stride = _pair(stride)
        self.padding = _pair(padding)

        # in TORCH: OUT x IN x H x W
        self.weight = nn.Parameter(torch.randn((out_channels, in_channels, *self.kernel_size)))
        self.bias = nn.Parameter(torch.randn(out_channels))

    def forward(self, x):
        bz, C, H, W = x.shape
        h_pad, w_pad = self.padding
        x = nn.functional.pad(x, (h_pad, h_pad, w_pad, w_pad), value=0)

        h_ker, w_ker = self.kernel_size
        h_stride, w_stride = self.stride

        h_in, w_in = x.shape[-2:]

        h_out = get_out_size(h_in, h_ker, h_stride)
        w_out = get_out_size(w_in, w_ker, w_stride)

        res = torch.zeros((bz, self.out_channels, h_out, w_out))
        for b_ind in range(bz): # перебираем индексы батчей
            for out_ind in range(self.out_channels): # перебираем индексы фильтры
                for c_ind in range(C): # перебираем индексы входных каналов
                    ker = self.weight.data[out_ind][c_ind] # текущее ядро
                    img = x[b_ind][c_ind] # текущее одноканальное изображение
                    for i in range(h_out): # height-индекс выходного изображения
                        for j in range(w_out): # width-индекс выходного изображения
                            start_img_h_pos = i * h_stride
                            start_img_w_pos = j * w_stride
                            # start_img_h_pos, start_img_w_pos - левый верхний индекс позиции ядра на изображении
                            for a in range(h_ker): # height-индекс ядра
                                for b in range(w_ker): # width-индекс ядра
                                    img_h_pos = start_img_h_pos + a
                                    img_w_pos = start_img_w_pos + b
                                    cur_img_value = img[img_h_pos][img_w_pos]
                                    cur_ker_value = ker[a][b]
                                    res[b_ind][out_ind][i][j] += cur_img_value * cur_ker_value
                res[b_ind][out_ind] += self.bias.data[out_ind]
        return res
```

```python
in_channels = 2
rows = 3
cols = 4
out_channels = 3
kernel_size = (2, 2)
stride = (1, 2)
padding = 1
```

```python
torch_conv = nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding)
my_conv = BrokenConv2d(in_channels, out_channels, kernel_size, stride, padding)
my_conv.weight.data = torch_conv.weight.data
my_conv.bias.data = torch_conv.bias.data
```

```python
bs = 2
input = torch.randn((bs, in_channels, rows, cols))
```

```python
my_res = my_conv(input)
my_res
```

```python
torch_res = torch_conv(input)
torch_res
```

```python
grad = torch.randn(my_res.shape)
```

```python
torch_res.backward(grad)
```

```python
my_res.backward(grad)
```
